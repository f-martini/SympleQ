```python
def invert_gate(G):
    g_name = G.name
    g_dim = G.dimension
    g_qudit_indices = G.qudit_indices
    g_phase_vector = G.phase_vector
    g_symplectic = np.array(G.images)
    Omega = quaos_random._symplectic_form(int(g_symplectic.shape[0]/2))
    g_inv_symplectic = (Omega.T @ g_symplectic.T @ Omega)%2
    g_inv_symplectic = g_inv_symplectic.tolist()
    g_inv_symplectic = [np.array(v) for v in g_inv_symplectic]
    G_inv = Gate(
        name=g_name + "_inv",
        dimension=g_dim,
        qudit_indices=g_qudit_indices,
        phase_vector=g_phase_vector,
        images=g_inv_symplectic
    )
    return(G_inv)
```

```python
def random_sym_Pauli(G, n_qudits=None, n_paulis=None):
    if n_qudits is None:
        n_qudits = G.dimension + 1
    if n_paulis is None:
        n_paulis = 2* n_qudits  
    P = quaos_random.pauli_hamiltonian(n_qudits,n_paulis,0,0)
    G_inv = invert_gate(G)
    P_prime = G_inv.act(P)
    P_sym = P + P_prime
    return(P_sym)
        
```

```python
def symplectic_form(n):
    """Return the standard 2n x 2n symplectic form matrix."""
    I = np.identity(n, dtype=np.uint8)
    O = np.zeros((n, n), dtype=np.uint8)
    return np.block([
        [O, I],
        [I, O]
    ])

def symplectic_dot(u, v, J):
    """Symplectic inner product: u^T J v mod 2"""
    return int(np.dot(u, J @ v) % 2)

def random_isotropic_vector(J):
    """Generate a random vector v such that <v, v> = 0."""
    i = 0
    while True:
        print(i)
        i +=1
        v = np.random.randint(0, 2, size=(J.shape[0],), dtype=np.uint8)
        if symplectic_dot(v, v, J) == 0:
            return v
```

```python
def random_cliffords(tableau, n_gates):
    _, _, qubits = tableau.shape
    gates = np.random.choice(CLIFFORDS, int(n_gates))

    for i in range(n_gates):
        # generate a random array of 2-tuples with c_gate and optional t_gate
        q = np.random.choice(qubits, 2, replace=False)
        gates[i] = gates[i](q[0], q[1])

        # apply gate
        tableau = gates[i](tableau)

    return tableau, gates


def synthetic_symmetries(qubits, r, c):
    symm_row= 2*r+c

    tableau_s = np.zeros([2, symm_row, qubits], dtype='int')
    np.fill_diagonal(tableau_s[0, :r, :r], 1)
    np.fill_diagonal(tableau_s[1, r:2*r, :r], 1)
    np.fill_diagonal(tableau_s[1, 2*r:, r:r+c], 1)

    return tableau_s


def commute(tableau_a, tableau_b):
    """
    returns True if every term in tableau_a commutes
    with every term in tableau_b.
    """
    if (len(tableau_a.shape) < 3) or (len(tableau_b.shape) < 3):
        return True

    _, r, c = tableau_a.shape

    for i in range(r):
        xor_ab = np.logical_xor(tableau_a[:, i, :].reshape(2, 1, c), tableau_b) # false if both are same paulis
        and_ab = np.logical_and(tableau_a[:, i, :].reshape(2, 1, c), tableau_b) # false if 0s
        out = np.logical_and(and_ab, xor_ab)
        f = out.sum()
        if f != 0:
            return False

    return True
```
